```python
import pandas as pd
```

```python
#reading the csv file
df_population = pd.read_csv('population_total.csv')
```

```python
df_population.dropna(inplace = True)
```

```python
# creating a pivot table
df_pivot = df_population.pivot(index='year', columns='country',
                                        values='population')
```

```python
df_pivot
```

```python
#selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 
                          'Indonesia', 'Brazil']]
```

```python
df_pivot
```

```python
df_pivot.plot(kind = 'line',xlabel='Years', ylabel='Population',
              title='Population (1955-2020)',
              figsize=(10,4))
```

```python
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
df_pivot_2020 = df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind='bar', color='orange',
                   xlabel='Years', ylabel='Population',
                   title='Population in 2020')
```

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_sample.plot(kind='bar',
                     xlabel='Years', ylabel='Population',
                     figsize=(8,4))
```
